```python
# Exercise 3
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
# Exercise 4
ages = np.array([22, 34, 18, 27, 45])
engagement_rates = np.array([0.25, 0.76, 0.58, 0.91, 0.47])
users = np.array(["Alex", "Jamie", "Morgan", "Taylor", "Jordan"])

print(ages)
print(engagement_rates)
print(users)
```

```python
# Exercise 4 continued
print(users[2])
print(np.sum(ages))
print(np.mean(engagement_rates))
```

```python
# Exercise 4 continued
users[0] = "Gizmo"
ages[-1] = 50

print(users)
print(ages)
```

```python
# Exercise 5
df = pd.DataFrame({
    "Name": users,
    "Age": ages,
    "Engagement": engagement_rates
})

print(df)
```

Double click and enter answer here: Structured.

```python
# Exercise 5 continued
print(df['Age'].mean())
print(df['Engagement'].mean())
print(df['Engagement'].median())
print(df['Engagement'].mode())
print(df['Engagement'].mode()[0])
```
